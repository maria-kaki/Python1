# Pseudocódigo 2

1. Inicio
2. Insira os dias, horas, minutos e segundos
3. Leia dias, horas, minutos e segundos
4. total = (dia*24*60*60) + (hora * 60*60) + (minutos *60) + segundos
5. Exibe total
6. Fim




```python
dia = int(input("Insira a quantidade de dias: "))
hora = int(input("Insira a quantidade de horas: "))
minutos = int(input("Insira a quantidade de minutos: "))
segundos = int(input("Insira os segundos: "))

total = (dia*24*60*60) + (hora * 60*60) + (minutos *60) + segundos

print("O valor total é  ",total," segundos")

```

# Pseudocódigo 3
1. Inicio
2. Insira o preco
3. Insira o desconto
4. Leia o preco e o desconto
5. total = preco - (preco * (desconto/100))
6. Mostre o total a ser pago e o percentual de desconto
7. Fim



```python
preco = float(input("Insira o preço: "))
desconto = float(input("Insira o desconto: "))

total = preco - (preco * (desconto/100))

print("O valor final é ", total)
print("O valor do desconto é de ", desconto, "%")
```

# Pseudocódigo 4

1. Inicio
2. Insira distancia
3. Insira velocidade
4. Leia distancia e velocidade
5. tempo = distancia/velocidade
6. Mostre tempo
7. Fim



```python
distancia = float(input("Insira a distancia: "))
velocidade = float(input("Insira a velocidade: "))

tempo = distancia/velocidade

print("O tempo de viagem é ", tempo)
```

# Problema 5

```python
print("Você deseja inserir:\n1)Tensão e Corrente\n2)Resistência e Corrente\n3)Tensão e Resistência")
opção = int(input())
if opção == 1:
    v = float(input('Insira a tensão em V: '))
    i = float(input('Insira o valor da corrente em A: '))
    p = v*i
    print("O valor da potência é de ",p," watts.")


elif opção == 2:
    r = float(input('Insira o valor da resistência em Ω: '))
    i = float(input('Insira o valor da corrente em A: '))
    p = r*i**2
    print("O valor da potência é de ",p," watts.")

elif opção == 3:
    v = float(input('Insira a tensão em V: '))
    r = float(input('Insira o valor da resistência em Ω: '))
    p = (v**2)/r
    print("O valor da potência é de ",p," watts.")
    
else:
    print('Insira um valor válido!')
```
