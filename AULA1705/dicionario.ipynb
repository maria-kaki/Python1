##### Crie uma função para contar até n (com while).

```python


def contar_n(n):
    """Função que conta até n"""

    i = 1;
    while i <=n:
        print(i)
        i +=1;

numero = int(input("Write a number for n!"));

contar_n(numero)
```

##### Crie uma função para contar até n (com for).

```python
def contar(n):
    for i in range(n):
        print(i+1);

numero = int(input("Write a number for n!"));
contar(numero)
```

##### Crie uma função para verificar cores de uma lista_cores = ['azul', 'verde','amarelo']

```python
def mostrar_cores():

    lista_cores = ['azul', 'verde','amarelo']  
    for i in lista_cores:
        print(i)

mostrar_cores()
```
